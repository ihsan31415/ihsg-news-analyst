***parsing dari tanggal naratif ke bentuk datetime yang ditentukan***

````python
import pandas as pd
from datetime import datetime
from dateutil import parser
import re

bulan_map = {
    "Januari": "01", "Februari": "02", "Maret": "03", "April": "04",
    "Mei": "05", "Juni": "06", "Juli": "07", "Agustus": "08",
    "September": "09", "Oktober": "10", "November": "11", "Desember": "12",
    "Jan": "01", "Feb": "02", "Mar": "03", "Apr": "04",
    "Mei": "05", "Jun": "06", "Jul": "07", "Agu": "08",
    "Sep": "09", "Okt": "10", "Nov": "11", "Des": "12",
    "January": "01", "February": "02", "March": "03", "April": "04",
    "May": "05", "June": "06", "July": "07", "August": "08",
    "September": "09", "October": "10", "November": "11", "December": "12",
    "Jan": "01", "Feb": "02", "Mar": "03", "Apr": "04", "May": "05", "Jun": "06",
    "Jul": "07", "Aug": "08", "Sep": "09", "Oct": "10", "Nov": "11", "Dec": "12"
}

def parse_localized_date(text):
    try:
        # Hapus hari di awal (misal "Senin,")
        text = re.sub(r"^[A-Za-z]+,\s*", "", text)
        text = text.replace("WIB", "").strip()
        # Regex yg lebih fleksibel, bisa menangani tambahan whitespace
        match = re.match(r"(\d{1,2})\s+(\w+)\s+(\d{4})\s+(\d{2}):(\d{2})", text)
        if not match:
            raise ValueError("Format tanggal lokal tidak cocok")
        day, month_str, year, hour, minute = match.groups()
        month = bulan_map.get(month_str)
        if not month:
            raise ValueError(f"Bulan tidak dikenal: {month_str}")
        return datetime.strptime(f"{day}-{month}-{year} {hour}:{minute}", "%d-%m-%Y %H:%M")
    except Exception as e:
        print(f"Gagal parsing lokal: '{text}' -> {e}")
        return None

def parse_any_date(text):
    try:
        # coba dateutil dulu
        return parser.parse(text, dayfirst=True)
    except Exception:
        dt = parse_localized_date(text)
        if dt is not None:
            return dt
        # fallback default date kalau gagal semua
        # bisa diganti sesuai kebutuhan
        fallback = datetime(1900,1,1,0,0)
        print(f"Parsing gagal total untuk: '{text}', return fallback {fallback}")
        return fallback
````

````python
df3["datetime"] = df3["datetime"].apply(parse_any_date)
````

# **SCRAPING**

````python
import requests
from bs4 import BeautifulSoup
from dateutil import parser

# Step 1: Fetch the article
url = "https://finansial.bisnis.com/read/20230309/90/1635538/bank-panin-dubai-syariah-pnbs-cetak-laba-rp25053-miliar-sepanjang-2022"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Step 2: Try to extract the publication date from meta tags
meta_date = soup.find("meta", {"name": "article:published_time"}) or \
            soup.find("meta", {"property": "article:published_time"})

if meta_date:
    pub_date = parser.parse(meta_date["content"])
    print("Publication date:", pub_date)
else:
    # Fallback: Look for visible date string in HTML elements (specific to bisnis.com)
    date_span = soup.find("div", class_="author")  # publication time is often here
    if date_span:
        text = date_span.get_text()
        try:
            pub_date = parser.parse(text, fuzzy=True, dayfirst=True)
            print("Publication date (fallback):", pub_date)
        except:
            print("Could not parse fallback date.")
    else:
        print("Publication date not found.")
````

***title scraper --verbose***

````python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def extract_title(url):
    print(f"\nProcessing URL: {url}")
    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    try:
        print("Sending HTTP GET request...")
        response = requests.get(url, headers=headers, timeout=10, allow_redirects=True)
        response.raise_for_status()
        print("Request successful, parsing HTML content...")
        soup = BeautifulSoup(response.text, 'html.parser')

        # Bisnis.com
        if "bisnis.com" in url:
            print("Detected site: bisnis.com")
            title_tag = soup.find("meta", property="og:title") or soup.find("title")
            if title_tag:
                title = title_tag["content"] if title_tag.has_attr("content") else title_tag.get_text()
                print(f"Found title: {title}")
                return title.strip()

        # CNBC Indonesia
        elif "cnbcindonesia.com" in url:
            print("Detected site: cnbcindonesia.com")
            title_tag = soup.find("meta", property="og:title") or soup.find("title")
            if title_tag:
                title = title_tag["content"] if title_tag.has_attr("content") else title_tag.get_text()
                print(f"Found title: {title}")
                return title.strip()

        else:
            print("URL domain not handled by this function.")

    except requests.exceptions.RequestException as req_err:
        print(f"Request error for {url}: {req_err}")
    except Exception as e:
        print(f"General error processing {url}: {e}")

    print(f"Returning None for {url} (no title found or error occurred)")
    return None
````

***date scraper --verbose***

````python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from dateutil import parser

def extract_pub_date(url):
    print(f"\nProcessing URL: {url}")
    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    try:
        print("Sending HTTP GET request...")
        response = requests.get(url, headers=headers, timeout=10, allow_redirects=True)
        response.raise_for_status()
        print("Request successful, parsing HTML content...")
        soup = BeautifulSoup(response.text, 'html.parser')

        # Bisnis.com
        if "bisnis.com" in url:
            print("Detected site: bisnis.com")
            meta_date = soup.find("meta", {"name": "article:published_time"}) or \
                            soup.find("meta", {"property": "article:published_time"})
            if meta_date:
                print(f"Found meta tag with date content: {meta_date['content']}")
                parsed_date = parser.parse(meta_date["content"])
                print(f"Parsed date: {parsed_date}")
                return parsed_date
            author_div = soup.find("div", class_="author")
            if author_div:
                text = author_div.get_text()
                print(f"Found author div text: {text}")
                parsed_date = parser.parse(text, fuzzy=True, dayfirst=True)
                print(f"Parsed date from author div: {parsed_date}")
                return parsed_date
            print("No date found for bisnis.com in meta or author div.")

        # CNBC Indonesia
        elif "cnbcindonesia.com" in url:
            print("Detected site: cnbcindonesia.com")
            meta_date = soup.find("meta", {"property": "article:published_time"})
            if meta_date:
                print(f"Found meta tag with date content: {meta_date['content']}")
                parsed_date = parser.parse(meta_date["content"])
                print(f"Parsed date: {parsed_date}")
                return parsed_date
            date_div = soup.find("div", class_="text-cm text-gray")
            if date_div:
                text = date_div.get_text()
                print(f"Found date div text: {text}")
                try:
                    parsed_date = parser.parse(text, fuzzy=True, dayfirst=True)
                    print(f"Parsed date from div: {parsed_date}")
                    return parsed_date
                except ValueError:
                    print(f"Could not parse date from: {text}")
            print("No date found for cnbcindonesia.com in meta or the specified div.")

        else:
            print("URL domain not handled by this function.")

    except requests.exceptions.RequestException as req_err:
        print(f"Request error for {url}: {req_err}")
    except Exception as e:
        print(f"General error processing {url}: {e}")

    print(f"Returning None for {url} (no date found or error occurred)")
    return None
````

````python
from sentence_transformers import SentenceTransformer

labse = SentenceTransformer("sentence-transformers/LaBSE")
````

````python
def clean_and_embed(text):
    try:
        cleaned = str(text).replace('\xa0', ' ').strip()
        return labse.encode(cleaned)
    except Exception as e:
        print(f"Failed on text: {text[:50]}... | Error: {e}")
        return None
````

***df2***

````python
from tqdm import tqdm
tqdm.pandas()
df2['embedding_isi'] = df2['isi'].progress_apply(clean_and_embed)
````
